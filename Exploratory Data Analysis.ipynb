Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Load The Data


```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

preprocessing for data and null values

```python
df = df.drop(["Country","Postal Code"],axis=1)
print(df.isnull().values.any())
df.head()
```

Choose the Style Colors

```python
plt.style.use('dark_background')
```

Plot Ship Mode with their Sales after that with their Profit

```python
plt.figure(figsize=(10,5))
plt.title("Ship mode and their sales")
sns.barplot(x = df['Ship Mode'] , y = df["Sales"])
```

```python
plt.figure(figsize=(10,5))
plt.title("Ship mode and their profit")
sns.barplot(x = df['Ship Mode'] , y = df["Profit"])
```

Segements with their Sales

```python
plt.figure(figsize=(10,5))
plt.title("Segment and the Sales")
sns.barplot(x = df['Segment'] , y = df['Sales'])
```

Segment With Their Profits

```python
plt.figure(figsize=(10,5))
plt.title("Segment and the Profit")
sns.barplot(x = df['Segment'] , y = df['Profit'])
```

Region With The Sales

```python
plt.figure(figsize=(10,5))
plt.title("Region and the Sales")
sns.barplot(x = df['Region'] , y = df['Sales'])
```

Region With Profit

```python
plt.figure(figsize=(10,5))
plt.title("Region and the Profit")
sns.barplot(x = df['Region'] , y = df['Profit'])
```

```python
plt.figure(figsize=(10,5))
plt.title("Region and Discount")
sns.barplot(x = df['Region'] , y = df['Discount'])
```

State With Sales

```python
plt.figure(figsize=(10,5))
plt.title("State and the Sales")
sns.barplot(x = df['State'] , y = df['Sales'])
plt.xticks(rotation=90)
```

State With Profit

```python
plt.figure(figsize=(10,5))
plt.title("State and the Profit")
sns.barplot(x = df['State'] , y = df['Profit'])
plt.xticks(rotation=90)
```

State With Their Oreders

```python
plt.figure(figsize=(10,5))
plt.title("Number of orders for each State")
sns.countplot(df['State'],label="Count")
plt.xticks(rotation=90)
```

The Category With the Order

```python
plt.figure(figsize=(10,5))
plt.title("Number of orders for each Category")
sns.countplot(df['Category'],label="Count")
```

The Category With the Sales

```python
plt.figure(figsize=(10,5))
plt.title("Sales for each Category")
sns.barplot(x=df['Category'],y=df['Sales'])
```

The Category With the Profit

```python
plt.figure(figsize=(10,5))
plt.title("Profit for each Category")
sns.barplot(df['Category'],df['Profit'])
```

Sub Category With the Orders

```python
plt.figure(figsize=(10,5))
plt.title("Number of orders for each Sub-Category")
sns.countplot(df['Sub-Category'],label="Count")
plt.xticks(rotation=90)
```

Sub-Category With the Sales

```python
plt.figure(figsize=(10,5))
plt.title("Sales for each Sub-Category")
sns.barplot(x=df['Sub-Category'],y=df['Sales'])
plt.xticks(rotation=90)
```

Sub-Category With The Profits

```python
plt.figure(figsize=(10,5))
plt.title("Profits for each Sub-Category")
sns.barplot(x=df['Sub-Category'],y=df['Profit'])
plt.xticks(rotation=90)
```
